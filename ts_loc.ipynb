```python
cat_feats = cPickle.load(open('cat_feats.p', 'rb')) + ['interest_level']

for col in cat_feats:
    df_all[col] = df_all[col].astype('category')
```

```python
df_all = df_all[(df_all['longitude'] > -80) & (df_all['longitude'] < 0)]
```

```python
x = df_all.drop(['interest_level'], 1)
y = df_all[['interest_level']]
```

```python
drop_list = [ u'features', u'listing_id', 'index']

x.drop(drop_list, axis=1, inplace=True)
```

```python
x['longitude'] = x['longitude'].apply(lambda x: abs(x))
```

```python
x['longitude'] = x['longitude'].apply(lambda x: round(x, 3))
x['latitude'] = x['latitude'].apply(lambda x: round(x, 3))
```
